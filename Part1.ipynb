```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
from langdetect import detect
import numpy as np
import langid
import datetime
import time
```

```python
data = pd.read_csv('/Users/ywj/Desktop/summerproject/Shanghai/2022.3/SH-review.csv') 
x = data['comments'].values.tolist()
language = []
def det(x):
    for i in range(0,len(x)):    
        try:
            lang = langid.classify(x[i])
        except:
            lang = 'Other'    # languages of some record can not be detected, so replace it with 'Other' 
        return lang

data['Lang'] = data['comments'].apply(det)

language = []
for i in range(0, len(x)):
    language.append(data['Lang'][i][0])

language
```

```python
from googletrans import Translator
import pandas as pd   

text = pd.read_csv('/Users/ywj/Desktop/summerproject/Shanghai/2022.3/Chinese copy 2.csv', header = None)
headers = ['comments', 'translation']
translator = Translator()
df = pd.DataFrame(index=range(0,len(text)), columns=headers)


def translate(row):
    ''' translate elements A and B within `row`. '''
    m = translator.translate(row[0], dest='En')
    return pd.Series([m.origin, m.text], headers)


for i, row in enumerate(text.values):
    # Fill empty dataframe with given serie.
    df.loc[i] = translate(row)

print(df)
```

```python
import string
from string import digits
data = pd.read_csv('/Users/ywj/Desktop/summerproject/Shanghai/2022.3/Chinese-SH copy .csv', header = None)
m = data[0].values.tolist()   # we did not set header name so the default value for header is 0, can change to 1
n = []
for i in range(len(m)):
    old_str = m[i]
    mid_val = old_str.encode("ascii", "ignore")        # remove all the non-ASCII characters
    string1 = mid_val.decode()
    table = str.maketrans('','',digits)
    new_str = string1.translate(table)                 # remove all the numbers
    new_str = new_str.replace('<br/>', '')
    new_str = new_str.replace('+','')
    new_str = new_str.translate(str.maketrans('','',string.punctuation))  # remove all the punctuations
    new_str = new_str.lower() # turn upper class into lower class
    n.append(new_str)
n
df = pd.DataFrame(n)    
df
```
